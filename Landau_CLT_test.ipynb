```python
from scipy.stats import moyal, norm
import numpy as np
import matplotlib.pyplot as plt

from jupyterthemes import jtplot 
jtplot.style(context='notebook', fscale=1.4, spines=False, gridlines='--') 
```

```python
def CLT_test(sample_size, n_samples=10000):
    """
    This function takes the no. of RVs in a sample i.e. sample size
    and plots a histogram vs the expected gaussian distribution 
    of the sample means.
    """
    
    # Taking sample from the Landau distribution 
    sample = moyal.rvs(size=(n_samples,sample_size))
    sample_mean = np.mean(sample, axis=1)
    
    # Expected Gaussian ditribution metrics
    dist_mean = moyal.mean()
    dist_std = moyal.std()/np.sqrt(sample_size)
    gaussian = norm(dist_mean, dist_std)
    xplot = np.linspace(dist_mean - 4*dist_std, dist_mean + 4*dist_std, 10000)
    
    # Plotting the results
    fig, ax = plt.subplots()
    ax.plot(xplot, gaussian.pdf(xplot), color='r')
    ax.hist(sample_mean, bins=100, density=True, stacked=True)
    plt.show()
```

```python
CLT_test(50)
```
